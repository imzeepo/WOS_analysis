```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/Users/jiyounglim/Documents/study/공종설 네트워크 분석/output/stats_temp.xlsx')
```

```python
df['percentage_of_authors'] = df['num_authors_in_main_component']/df['num_community']*100
```

```python
df.columns
```

```python
# Assuming your DataFrame is named 'df'
x = df['degree_centrality_mean']
y = df['percentage_of_authors']
colors = df['field']
sizes = df['num_authors']*100  # Use 'num_authors' for dot sizes

# Set the style for seaborn
sns.set(style='whitegrid')

# Create the scatter plot
scatterplot = sns.scatterplot(x=x, y=y, hue=colors, size=sizes, palette='viridis', alpha=0.7, legend='brief')

# Set x and y labels
scatterplot.set(xlabel='Degree Centrality Mean', ylabel='Percentage of Authors')


# Show the plot
plt.show()
```
